```python
# 25. Infobox
# Extract field names and their values in the Infobox “country”, 
# and store them in a dictionary object.

def extract_jpwiki():
    fname = 'data/ch3/jawiki-country.json.gz'
    df = pd.read_json(fname, lines=True)
    uk_text = df.query('title=="イギリス"')['text'].values[0]
    return uk_text

uk_text = extract_jpwiki()

# format: {{基礎情報 国 .......}}
infobox = re.compile(r'''
    ^\{\{基礎情報  # Infobox starts
    (.*?)          # words after infobox
    ^\}\}$         # infobox end with }}
    ''', re.MULTILINE+re.VERBOSE+re.DOTALL)


# format: |公式国名 = {{lang|en|United Kingdom of Great Britain and Northern Ireland}}<ref>英語以外での正式国名:<br />
p = re.compile(r'''
    ^\|         # start from |
    (.+?)       # words after |
    \s*         # space after words
    =
    \s*         # space after words    
    (.+?)       # words after =
    \n
    ''', re.MULTILINE + re.VERBOSE + re.DOTALL)

contents = infobox.findall(uk_text)
fields = p.findall(contents[0])

print(f'''
show first 10 items
58 items in total
''')
for field in fields:
    print(field)
```

```python
'''26. Remove emphasis markups
In addition to the process of the problem 25,
remove emphasis MediaWiki markups from the values. 
See Help:Cheatsheet.'''

'''27. Remove internal links
In addition to the process of the problem 26,
remove internal links from the values.
See Help:Cheatsheet.'''

'''28. Remove MediaWiki markups
In addition to the process of the problem 27, 
remove MediaWiki markups from the values as much as you can,
and obtain the basic information of the country in plain text format.'''

def extract_jpwiki():
    fname = 'data/ch3/jawiki-country.json.gz'
    df = pd.read_json(fname, lines=True)
    uk_text = df.query('title=="イギリス"')['text'].values[0]
    return uk_text

def remove_markup(target):

    # delete '''
    pattern = re.compile(r'''
        (\'{2,5})   # delete 2~5 '
        (.*?)       # any words after '
        ''', re.MULTILINE + re.VERBOSE)
    target = pattern.sub(r'', target)
    

    #delete lang {{lang|tag|word}}
    pattern = re.compile(r'''\{\{lang\| #start from {{lang|
                         (?:            #things to be ignored
                          .+?           #tag words
                          \|)           #end of tag|
                         .*?         #match words after tag|
                         ''', re.MULTILINE + re.VERBOSE)
    target = pattern.sub(r'', target)
    
    # delete link
    target = re.sub(r'url\S+ |\[http\S+','', target)    

    # delete [],{}
    target = re.sub(r'(\[\[)|(\]\])|({{)|(}})','', target)  
    
    # delete <br> <ref>
    pattern = re.compile(r'''
        <           # start from <
        \/?         # \ appear once or not <ref><\ref>
        [br|ref]    # br or ref
        [^>]*?      # match words except > non-greedy
        >           # end with >
        ''', re.MULTILINE + re.VERBOSE)
    target = pattern.sub('', target)
    
    return target

infobox = re.compile(r'''
    ^\{\{基礎情報.*?$   # ' start from {{基礎情報
    (.*?)       # match words non-greedy
    ^\}\}$      # end with }}
    ''', re.MULTILINE + re.VERBOSE + re.DOTALL)


pattern = re.compile(r'''
    ^\|         # start from \n
    \s*
    (.+?)       # words after \n
    \s*         # space after words
    =
    \s*         # space after words    
    (.+?)       # words after =
    \n
    ''', re.MULTILINE + re.VERBOSE + re.DOTALL)


uk_text = extract_jpwiki()
contents = infobox.findall(uk_text)
fields = pattern.findall(contents[0])

result = {}
for field in fields:
    result[field[0]] = remove_markup(field[1])

    
print(f'''
show first 10 items
58 items in total
''')

for k in result.items():
    print(k)
```
